```python
image_path = "/content/wall.png"
original_image1 = cv2.imread(image_path)
original_image_rgb1 = cv2.cvtColor(original_image1, cv2.COLOR_BGR2RGB)

plt.imshow(original_image_rgb1)
plt.title("Original Image1")
#plt.axis("off")
plt.show()
```

```python
# Step 4: Apply image filtering
#3x3 Gaussian blur filter
kernel_size = (3, 3)
filtered_image1 = cv2.GaussianBlur(original_image, kernel_size, 0)
# Step 5: Display the filtered image
filtered_image_rgb1 = cv2.cvtColor(filtered_image1, cv2.COLOR_BGR2RGB)

plt.imshow(filtered_image_rgb1)
plt.title("Filtered Image")
#plt.axis("off")
plt.show()
```

```python
#Step 4 medianBlur
ksize= 1
filtered_image = cv2.medianBlur(original_image, ksize)
# Step 5: Display the filtered image
filtered_image_rgb = cv2.cvtColor(filtered_image, cv2.COLOR_BGR2RGB)

plt.imshow(filtered_image_rgb)
plt.title("Filtered Image")
#plt.axis("off")
plt.show()
```

```python
# Apply the bilateral filter
d = 9   # Diameter of each pixel neighborhood used during filtering (larger value means more distant pixels will influence the filtering)
sigma_color = 75   # Filter sigma in the color space (larger value means colors further away from the central pixel will be mixed)
sigma_space = 75   # Filter sigma in the coordinate space (larger value means pixels further away from the central pixel will influence the filtering)

filtered_image = cv2.bilateralFilter(original_image, d, sigma_color, sigma_space)

# Step 5: Display the filtered image
filtered_image_rgb = cv2.cvtColor(filtered_image, cv2.COLOR_BGR2RGB)

plt.imshow(filtered_image_rgb)
plt.title("Filtered Image")
#plt.axis("off")
plt.show()
```

```python
#Edge Detection
image_path = "/content/wall.png"
original_image = cv2.imread(image_path)
original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)

plt.imshow(original_image_rgb)
plt.title("Original Image")
#plt.axis("off")
plt.show()
```

```python
# Step 3: Apply edge detection (Canny edge detection)
edges = cv2.Canny(original_image, threshold1=100, threshold2=200)

# You can experiment with different threshold values to control the edge detection results

# Step 4: Display the edges
plt.imshow(edges, cmap='gray')
plt.title("Edge Detection (Canny)")
#plt.axis("off")
plt.show()
```

```python
image_path = "/content/Duraga Ma.jpg"
original_image2 = cv2.imread(image_path)
original_image_rgb2 = cv2.cvtColor(original_image2, cv2.COLOR_BGR2RGB)

plt.imshow(original_image_rgb2)
plt.title("Original Image2")
#plt.axis("off")
plt.show()
```

```python
# Step 3: Apply image enhancement (Histogram Equalization)
gray_image = cv2.cvtColor(original_image2, cv2.COLOR_BGR2GRAY)
enhanced_image = cv2.equalizeHist(gray_image)

# Step 4: Display the enhanced image
plt.imshow(enhanced_image, cmap='gray')
plt.title("Enhanced Image (Histogram Equalization)")
plt.axis("off")
plt.show()
```
